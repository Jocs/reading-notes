# 数据分析案例

```python
import pandas as pd

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('datasets/movielens/users.dat',
                     sep='::', header=None, names=unames, engine='python')

users.head()
```

```python
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('datasets/movielens/ratings.dat',
                      sep='::', header=None, names=rnames, engine='python')
ratings.head()
```

```python
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('datasets/movielens/movies.dat',
                      sep='::', header=None, names=mnames, engine='python')
movies.head()
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
data.head()
```

```python
mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mean_ratings[:5]
```

```python
ratings_by_title = data.groupby('title').size()
ratings_by_title[:10]
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]

active_titles
```

```python
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings
```

```python
top_females_ratings = mean_ratings.sort_values(by='F', ascending=False)
top_females_ratings[:10]
```

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
mean_ratings[:5]
```

```python
sorted_by_diff = mean_ratings.sort_values(by='diff')
sorted_by_diff[:10]
```

```python
sorted_by_diff[::-1][:10]
```

```python
rating_std_by_title = data.groupby('title')['rating'].std()
rating_std_by_title[:10]
```

```python
rating_std_by_title.sort_values(ascending=False)[:10]
```

```python
movies['genres'].head().str.split('|')
```

```python
movies['genre'] = movies.pop('genres').str.split('|')
movies.head()
```

```python
movies_exploded = movies.explode('genre')
movies_exploded.head()
```

```python
ratings_with_genre = pd.merge(pd.merge(movies_exploded, ratings), users)
ratings_with_genre.head()
```

```python
genre_rating = ratings_with_genre.groupby(['genre', 'age'])['rating'].mean().unstack('age')
genre_rating.head()
```
